# Docker的远程访问

修改Docker守护进程启动选项：

-H tcp://host:port或unix:///path/to/socket或fd://socketfd

守护进程默认配置：-H unix:///var/run/docker.sock

例如我的/etc/default/docker配置文件如下：

```
# Docker Upstart and SysVinit configuration file

# Customize location of Docker binary (especially for development testing).
#DOCKER="/usr/local/bin/docker"

# Use DOCKER_OPTS to modify the daemon startup options.
#DOCKER_OPTS="--dns 8.8.8.8 --dns 8.8.4.4"

# If you need Docker to use an HTTP proxy, it can also be specified here.
#export http_proxy="http://127.0.0.1:3128/"

# This is also a handy place to tweak where Docker's temporary files go.
#export TMPDIR="/mnt/bigdrive/docker-tmp"
DOCKER_OPTS="--label name=ubuntu-1-docker -H tcp://0.0.0.0:2375 -H unix:///var/run/docker.sock"
```

----

修改客户端配置同守护进程一样：

```
docker -H tcp://192.168.42.131:2375 info
```

如果觉得每次访问都需要加-H参数很麻烦，可以配置DOCKER_HOST环境变量：

```
export DOCKER_HOST="tcp://192.168.42.131:2375"
```
